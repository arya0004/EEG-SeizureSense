```python
# ==========================================================
# CHB-MIT Preprocessing (Optimized, Memory-Safe Version)
# ==========================================================

import os
import numpy as np
import mne
from tqdm import tqdm
import gc

data_root = "/content/archive.physionet.org/pn6/chbmit"
patients = ["chb01","chb02","chb03","chb04","chb05"]

window_size = 4
overlap = 0.5

X, y, pids = [], [], []

def parse_summary(summary_path):
    seizure_dict = {}
    current_file = None

    with open(summary_path, "r") as f:
        lines = f.readlines()

    for line in lines:
        line = line.strip()

        if line.startswith("File Name:"):
            current_file = line.split(":")[1].strip()
            seizure_dict[current_file] = []

        if "Seizure Start Time" in line:
            start = float(line.split(":")[1].split()[0])

        if "Seizure End Time" in line:
            end = float(line.split(":")[1].split()[0])
            seizure_dict[current_file].append((start, end))

    return seizure_dict


for pid in patients:

    print("\nProcessing:", pid)

    patient_path = os.path.join(data_root, pid)
    summary_path = os.path.join(patient_path, f"{pid}-summary.txt")

    if not os.path.exists(summary_path):
        print("Summary file missing for", pid)
        continue

    seizure_info = parse_summary(summary_path)
    edf_files = sorted([f for f in os.listdir(patient_path) if f.endswith(".edf")])

    for edf_file in tqdm(edf_files):

        edf_path = os.path.join(patient_path, edf_file)

        try:
            raw = mne.io.read_raw_edf(edf_path, preload=False, verbose=False)

            # ✅ Updated API
            raw.pick([raw.ch_names[0]])

            raw.load_data()
            raw.filter(0.5, 40, verbose=False)

            data = raw.get_data().astype(np.float32)
            sfreq = raw.info['sfreq']

            samples = int(window_size * sfreq)
            step = int(samples * (1 - overlap))

            seizure_intervals = seizure_info.get(edf_file, [])

            for start in range(0, data.shape[1] - samples + 1, step):

                end = start + samples
                segment = data[:, start:end]

                segment = (segment - np.mean(segment)) / (np.std(segment) + 1e-8)

                start_sec = start / sfreq
                end_sec = end / sfreq

                label = 0

                for sz_start, sz_end in seizure_intervals:
                    overlap_sec = max(
                        0,
                        min(end_sec, sz_end) - max(start_sec, sz_start)
                    )

                    if overlap_sec >= window_size / 2:
                        label = 1
                        break

                X.append(segment)
                y.append(label)
                pids.append(pid)

            del raw
            gc.collect()

        except Exception as e:
            print("Error in", edf_file, "->", e)


X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.int64)
pids = np.array(pids)

print("\n====================================")
print("Final CHB Shape:", X.shape)
print("Total segments:", len(y))
print("Total seizure segments:", np.sum(y))
print("Seizure ratio:", np.mean(y))
print("====================================")

print("\nSeizure segments per patient:")
for pid in np.unique(pids):
    print(pid, "->", np.sum(y[pids == pid]))
```

```python
# ==========================================================
# Create Balanced CHB Dataset
# ==========================================================

# Get seizure and non-seizure indices
seizure_idx = np.where(y == 1)[0]
nonseizure_idx = np.where(y == 0)[0]

# Randomly sample equal non-seizure windows
np.random.seed(42)
nonseizure_sample = np.random.choice(
    nonseizure_idx,
    size=len(seizure_idx),
    replace=False
)

# Combine
balanced_idx = np.concatenate([seizure_idx, nonseizure_sample])

# Shuffle
np.random.shuffle(balanced_idx)

# Create balanced dataset
X_bal = X[balanced_idx]
y_bal = y[balanced_idx]
pids_bal = pids[balanced_idx]

print("Balanced shape:", X_bal.shape)
print("Seizure ratio:", np.mean(y_bal))
```

```python
# ==========================================================
# Bonn Dataset Preprocessing (Final Safe Version)
# ==========================================================

import os
import numpy as np
from scipy.signal import resample

bonn_root = "/content/bonn"

X_bonn, y_bonn = [], []

# Only process folders that actually exist
folders = [f for f in ["Z","O","N","F","S"] if os.path.exists(os.path.join(bonn_root, f))]

for folder in folders:

    folder_path = os.path.join(bonn_root, folder)

    for file in os.listdir(folder_path):

        if file.endswith(".txt"):

            full_path = os.path.join(folder_path, file)

            signal = np.loadtxt(full_path)

            # Resample to 1024 (match CHB window)
            signal = resample(signal, 1024)

            # Normalize (same as CHB)
            signal = (signal - np.mean(signal)) / (np.std(signal) + 1e-8)

            X_bonn.append(signal.astype(np.float32))
            y_bonn.append(1 if folder == "S" else 0)

X_bonn = np.array(X_bonn)[:, np.newaxis, :]
y_bonn = np.array(y_bonn, dtype=np.int64)

print("====================================")
print("Bonn shape:", X_bonn.shape)
print("Total seizure segments:", np.sum(y_bonn))
print("Seizure ratio:", np.mean(y_bonn))
print("====================================")
```

```python
import torch
import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super().__init__()

        # Convolution layers
        self.conv1 = nn.Conv1d(1, 32, kernel_size=5)
        self.pool = nn.MaxPool1d(2)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=5)

        # Dropout for better generalization
        self.dropout = nn.Dropout(0.3)

        # Fully connected layers
        self.fc1 = nn.Linear(64 * 253, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):

        # Conv block 1
        x = self.pool(torch.relu(self.conv1(x)))

        # Conv block 2
        x = self.pool(torch.relu(self.conv2(x)))

        # Flatten
        x = x.view(x.size(0), -1)

        # Dense + dropout
        x = self.dropout(torch.relu(self.fc1(x)))

        # IMPORTANT: no sigmoid here
        x = self.fc2(x)

        return x
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def train_eval(X_train, y_train, X_test, y_test, epochs=25, batch_size=32):

    model = CNN().to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.BCEWithLogitsLoss()

    # Convert to tensors
    X_train_t = torch.tensor(X_train, dtype=torch.float32)
    y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)

    train_dataset = TensorDataset(X_train_t, y_train_t)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # ---------------- TRAIN ----------------
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)

            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        if (epoch + 1) % 5 == 0:
            print(f"Epoch [{epoch+1}/{epochs}] Loss: {running_loss:.4f}")

    # ---------------- TEST ----------------
    model.eval()
    with torch.no_grad():
        X_test_t = torch.tensor(X_test, dtype=torch.float32).to(device)
        logits = model(X_test_t)
        probs = torch.sigmoid(logits).cpu().numpy()

    preds = (probs > 0.5).astype(int).flatten()

    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    cm = confusion_matrix(y_test, preds)

    print("Positive predictions:", np.sum(preds))
    print("Total predictions:", len(preds))

    return acc, f1, cm, model
```

```python
from sklearn.model_selection import train_test_split

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_bal,
    y_bal,
    test_size=0.2,
    stratify=y_bal,
    random_state=42
)

acc_r, f1_r, cm_r, model_r = train_eval(
    X_train_r,
    y_train_r,
    X_test_r,
    y_test_r
)

print("\n===== RANDOM SPLIT RESULTS =====")
print("Accuracy:", acc_r)
print("F1 Score:", f1_r)
print("Confusion Matrix:\n", cm_r)
```

```python
# ==========================================================
# Multi-Seed Evaluation
# ==========================================================

from sklearn.model_selection import train_test_split
import numpy as np

seeds = [0, 1, 2, 3, 4]

acc_random_list = []
acc_patient_list = []
acc_bonn_list = []

for seed in seeds:

    print("\nRunning seed:", seed)

    # ----- RANDOM SPLIT -----
    X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
        X_bal, y_bal,
        test_size=0.2,
        stratify=y_bal,
        random_state=seed
    )

    acc_r, f1_r, cm_r, model_r = train_eval(
        X_train_r, y_train_r,
        X_test_r, y_test_r
    )

    # ----- PATIENT-WISE -----
    test_patients = ["chb05"]

    train_mask = ~np.isin(pids_bal, test_patients)
    test_mask = np.isin(pids_bal, test_patients)

    acc_p, f1_p, cm_p, model_p = train_eval(
        X_bal[train_mask], y_bal[train_mask],
        X_bal[test_mask], y_bal[test_mask]
    )

    # ----- CROSS DATASET -----
    model_r.eval()
    with torch.no_grad():
        X_bonn_t = torch.tensor(X_bonn, dtype=torch.float32).to(device)
        logits = model_r(X_bonn_t)
        probs = torch.sigmoid(logits).cpu().numpy()

    preds_bonn = (probs > 0.5).astype(int).flatten()
    acc_bonn = accuracy_score(y_bonn, preds_bonn)

    acc_random_list.append(acc_r)
    acc_patient_list.append(acc_p)
    acc_bonn_list.append(acc_bonn)

print("\n=== AVERAGE RESULTS OVER SEEDS ===")
print("Random Avg:", np.mean(acc_random_list))
print("Patient Avg:", np.mean(acc_patient_list))
print("Bonn Avg:", np.mean(acc_bonn_list))
```

```python
# ==========================================================
# Statistical Testing
# ==========================================================

from scipy.stats import ttest_rel

t_val, p_val = ttest_rel(acc_random_list, acc_patient_list)

print("====================================")
print("Paired t-test (Random vs Patient)")
print("t-value:", t_val)
print("p-value:", p_val)
print("====================================")
```

```python
# ==========================================================
# PSD Plot (Seizure vs Non-Seizure)
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

sfreq = 256  # CHB sampling rate

# Separate classes
seizure_samples = X[y == 1]
nonseizure_samples = X[y == 0]

# Take small subset to avoid memory overload
seizure_samples = seizure_samples[:200]
nonseizure_samples = nonseizure_samples[:200]

def compute_avg_psd(data):
    psds = []
    for sample in data:
        f, Pxx = welch(sample[0], fs=sfreq, nperseg=256)
        psds.append(Pxx)
    return f, np.mean(psds, axis=0)

f_s, psd_s = compute_avg_psd(seizure_samples)
f_n, psd_n = compute_avg_psd(nonseizure_samples)

plt.figure(figsize=(8,5))
plt.semilogy(f_s, psd_s, label="Seizure")
plt.semilogy(f_n, psd_n, label="Non-Seizure")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density")
plt.title("Average PSD (Seizure vs Non-Seizure)")
plt.legend()
plt.show()
```

```python
# ==========================================================
# EEG Frequency Band Power Comparison
# ==========================================================

bands = {
    "Delta (0.5-4 Hz)": (0.5, 4),
    "Theta (4-8 Hz)": (4, 8),
    "Alpha (8-13 Hz)": (8, 13),
    "Beta (13-30 Hz)": (13, 30),
    "Gamma (30-40 Hz)": (30, 40)
}

def band_power(psd, freqs, band):
    low, high = band
    idx = np.logical_and(freqs >= low, freqs <= high)
    return np.mean(psd[idx])

band_powers_s = []
band_powers_n = []

for band in bands.values():
    band_powers_s.append(band_power(psd_s, f_s, band))
    band_powers_n.append(band_power(psd_n, f_n, band))

x = np.arange(len(bands))

plt.figure(figsize=(8,5))
plt.bar(x - 0.2, band_powers_s, width=0.4, label="Seizure")
plt.bar(x + 0.2, band_powers_n, width=0.4, label="Non-Seizure")
plt.xticks(x, bands.keys(), rotation=45)
plt.ylabel("Average Power")
plt.title("EEG Band Power Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```
